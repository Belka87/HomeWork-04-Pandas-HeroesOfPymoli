Age Demographics:
- Establish bins for ages
- Categorize the existing players using the age bins. Hint: use pd.cut()
- Calculate the numbers and percentages by age group
- Create a summary data frame to hold the results
- Optional: round the percentage column to two decimal points
- Display Age Demographics Table

```python

# Create the bins in which data will be held
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

# Create labels for this bins
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"] 

# Slice the data and place it into bins
purchase_data_df["Age Range"] = pd.cut(purchase_data_df["Age"], age_bins, labels = group_labels)

age_demographics = purchase_data_df.groupby("Age Range")
# Calculating precentages and numbers by age group

total_count = age_demographics["SN"].nunique()
total_players = purchase_data_df["SN"].nunique()
percentage_of_players = (total_count/total_players) *100


# Placing information into a data frame
summary_df = pd.DataFrame({"Total Count": total_count, "Percentage of Players": percentage_of_players})


# Use .map to format all the columns

summary_df["Percentage of Players"] = summary_df["Percentage of Players"].astype(float).map("{:.2f}%".format)
summary_df
```

Purchasing Analysis (Age):
- Bin the purchase_data data frame by age
- Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
- Create a summary data frame to hold the results
- Optional: give the displayed data cleaner formatting
- Display the summary data frame

```python
# Create the bins in which data will be held
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

# Create labels for this bins
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"] 

# Slice the data and place it into bins
purchase_data_df["Age Range"] = pd.cut(purchase_data_df["Age"], age_bins, labels = group_labels)
purchase_data_df.head()

# Calculating the purchase count,average purchase price, avgerage perchase total per person

total_count = purchase_data_df.groupby(["Age Range"]).sum()["Price"]
total_count

avg_price = round(purchase_data_df.groupby(["Age Range"]).mean()["Price"], 2)
avg_price

purchase_count = purchase_data_df.groupby(["Age Range"]).count()["Price"]
purchase_count

avg_per_person = round((total_count/purchase_count), 2)
avg_per_person

# Placing information into a data frame
age_purchase_df = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": avg_price, 
                   "Total Purchase Value": total_count, "Average Purchase Total Per Person": avg_per_person})
age_purchase_df

# Use .map to format the columns
age_purchase_df["Average Purchase Price"] = age_purchase_df["Average Purchase Price"]. map("${:.2f}".format)
age_purchase_df["Total Purchase Value"] = age_purchase_df["Total Purchase Value"]. map("${:.2f}".format)
age_purchase_df["Average Purchase Total Per Person"] = age_purchase_df["Average Purchase Total Per Person"]. map("${:.2f}".format)
age_purchase_df
```

Top Spenders:
- Run basic calculations to obtain the results in the table below
- Create a summary data frame to hold the results
- Sort the total purchase value column in descending order
- Optional: give the displayed data cleaner formatting
- Display a preview of the summary data frame

```python
purchase_data_df.head()
```

```python
# Create a GroupBy object based up on "SN"
sn_purchase_df = purchase_data_df.groupby(["SN"])
sn_purchase_df.head()

# Run basic calculation
sn_purchase_count = sn_purchase_df["Purchase ID"].count()
sn_purchase_count.head()

sn_purchase_avg = round(sn_purchase_df["Price"].mean(), 2)
sn_purchase_avg.head()

sn_purchase_total = round(sn_purchase_df["Price"].sum(), 2)
sn_purchase_total.head()

# Placing the information above into a DataFrame
top_spenders_df = pd.DataFrame({"Purchase Count": sn_purchase_count, "Average Purchase Price": sn_purchase_avg, 
                                "Total Purchase Value": sn_purchase_total})
top_spenders_df.head()

# Sorting data from highest to lowest value
top_spenders_df = top_spenders_df.sort_values(by = "Total Purchase Value", ascending = False)
top_spenders_df.head()

# Use .map to format the columns
top_spenders_df["Average Purchase Price"] = top_spenders_df["Average Purchase Price"].map("${:.2f}".format)
top_spenders_df["Total Purchase Value"] = top_spenders_df["Total Purchase Value"].map("${:.2f}".format)
top_spenders_df.head()
```

Most Popular Items:
- Retrieve the Item ID, Item Name, and Item Price columns
- Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
- Create a summary data frame to hold the results
- Sort the purchase count column in descending order
- Optional: give the displayed data cleaner formatting
- Display a preview of the summary data frame

```python
purchase_data_df.head()
```

```python
# Create a GroupBy object
popular_items_df = purchase_data_df.groupby(["Item ID", "Item Name"])

item_purchase_count = popular_items_df["Purchase ID"].count()
item_purchase_count.head()

item_purchase_avg = popular_items_df["Price"].mean()
item_purchase_avg.head()

item_purchase_total = popular_items_df["Price"].sum()
item_purchase_total.head()

# Placing information above into a Data Frame
popular_items_df = pd.DataFrame({"Purchase Count": item_purchase_count, "Item Price": item_purchase_avg,
                                 "Total Purchase Value": item_purchase_total})
popular_items_df.head()


# Use .map to format the columns
popular_items_df["Item Price"] = popular_items_df["Item Price"].map("${:.2f}".format)
popular_items_df["Total Purchase Value"] = popular_items_df["Total Purchase Value"].map("${:.2f}".format)
popular_items_df.head()

# Sorting data from highest to lowest value
popular_items_df = popular_items_df.sort_values(by = "Purchase Count", ascending = False)
popular_items_df.head()
```

Most Profitable Items:
- Sort the above table by total purchase value in descending order
- Optional: give the displayed data cleaner formatting
- Display a preview of the data frame

```python
# Create a GroupBy object
popular_items_df = purchase_data_df.groupby(["Item ID", "Item Name"])

item_purchase_count = popular_items_df["Purchase ID"].count()
item_purchase_count.head()

item_price = popular_items_df["Price"].mean()
item_price.head()

item_purchase_total = item_purchase_count*item_price
item_purchase_total.head()

# Placing information above into a Data Frame
most_profitable_item_df = pd.DataFrame({"Purchase Count": item_purchase_count, "Item Price": item_price,
                                 "Total Purchase Value": item_purchase_total})
most_profitable_item_df.head()

# Sorting data from highest to lowest value
most_profitable_item_df = most_profitable_item_df.sort_values(by = "Total Purchase Value", ascending = False)
most_profitable_item_df.head()


# Use .map to format the columns
most_profitable_item_df["Item Price"] = most_profitable_item_df["Item Price"].map("${:.2f}".format)
most_profitable_item_df["Total Purchase Value"] = most_profitable_item_df["Total Purchase Value"].map("${:.2f}".format)
most_profitable_item_df.head()
```
